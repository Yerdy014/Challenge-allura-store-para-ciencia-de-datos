### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

###**Graficos**

###Grafico de barras ingresos por tienda

```python
import matplotlib.pyplot as plt
# Lista de ingresos de cada tienda
ingresos = [1150880400, 1116343500, 1098019600, 1038375700]

#DataFrame
df_ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos": ingresos
})

print(df_ingresos)

#Grafico
df_ingresos.plot(x="Tienda", y="Ingresos", kind="barh", legend=False,color="Blue")
plt.title("Ingresos por tienda")
plt.xlabel("Ingresos")
plt.ylabel("Tiendas")
plt.xticks(rotation=0)
plt.show()
```

###Grafico ventas por categoria

```python
import matplotlib.pyplot as plt
categorias = [
    "Electrónicos",
    "Electrodomésticos",
    "Muebles",
    "Instrumentos musicales",
    "Deportes y diversión",
    "Juguetes",
    "Artículos para el hogar",
    "Libros"
]

ventas = [
    429493500.0,
    363685200.0,
    187633700.0,
    91299000.0,
    39290000.0,
    17995700.0,
    12698400.0,
    8784900.0
]

# Crear DataFrame
df_ventas = pd.DataFrame({
    "Categoría del Producto": categorias,
    "Ventas": ventas
})
#creacion del grafico

df_ventas.plot(
    x="Categoría del Producto",
    y="Ventas",
    kind="bar",
    legend=False,
    color="skyblue"
)

plt.title("Ventas por Categoría")
plt.xlabel("Categorías")
plt.ylabel("Ventas")
plt.tight_layout()
plt.show()
```

###Grafico opinión de los clientes


```python
import matplotlib.pyplot as plt

#data frame promedio opinion de los clientes por tienda
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [
    3.976685036032217,
    4.037303942348453,
    4.048325561678677,
    3.9957591178965224
]


df_calificaciones = pd.DataFrame({
    "Tienda": tiendas,
    "Calificación Promedio": calificaciones
})

#grafico de lineas
plt.figure(figsize=(8,5))
plt.plot(
    df_calificaciones["Tienda"],
    df_calificaciones["Calificación Promedio"],
    marker="o", linestyle="-", color="blue"
)

# Personalizacion de  gráfico
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(3.9, 4.1)  # ajustar eje Y para ver mejor las diferencias
plt.grid(True)

plt.show()

```
